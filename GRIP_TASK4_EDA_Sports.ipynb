# ***Exploratory Data Analysis***

```python
import matplotlib.pyplot as plt
%matplotlib inline
corr=cri_df.corr() 
plt.figure(figsize=(10,9))
sns.heatmap(corr,annot=True,cmap='Blues',linewidth=1,linecolor='Black') 
plt.xticks(rotation=90)
```

## match id have positive correlation with id and win by runs 

```python
sns.boxplot(x=cri_df["win_by_runs"],palette='rocket_r',order=cri_df["win_by_runs"].value_counts()[0:1].index)
plt.xticks(rotation=45)
```

# ***As you can see in above graph, that most of the winning runs were in between 0-20***

```python
cri_df.hist(figsize=(12,12),bins=50)
plt.show()
```

## ***In win by wickets you can see that none has won without taking wickets and in the ball column you can see that 8 ball delivered by bowler are less beacause of no wides or no-ball.***

```python
fig,ax=plt.subplots(figsize=(20,10))
ax.scatter(cri_df['batsman'].head(230),cri_df['win_by_runs'].head(230))
ax.set_xlabel('BM')
ax.set_ylabel('BMS')
plt.show()
```

### ***From the above diagram, we can say that S Dhawan has scored runs at higher level for match winning than others.***

```python
sns.lineplot(x='season',y='winner',label='Winner',data=cri_df)
plt.legend()
plt.show()
```

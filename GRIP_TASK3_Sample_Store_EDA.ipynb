# ***Exploratory Data Analysis(EDA)***

```python
plt.rcParams['figure.figsize']=(15,3)
plt.bar(loss_df['Sub-Category'],loss_df['Sales']);
plt.rcParams.update({'font.size':10});
plt.xlabel('Sub_Category');
plt.ylabel('Sales');
```

# ***Here, we observed that the Sales for Fasteners, Appliances, Furnishings, and Accessories is very low***

```python
plt.rcParams['figure.figsize']=(28,8)
plt.bar(ss_df['Sub-Category'],ss_df['Sales']);
plt.rcParams.update({'font.size':14});
plt.xlabel('Sub_Category');
plt.ylabel('Sales');
```

# ***When it comes to comparison in overall supermarket data, Fasteners, Labels, Furnishings, Art, paper, Envelopes, etc., sub-categories have very fewer sales, thats why it needs to be improved.***

```python
plt.figure(figsize=(15,5))
plt.title('Most Used')
ss_df['Category'].value_counts().head(10).plot.pie()
plt.show()
```

# ***Here we can see that the most used category from the three is Office Supplies***

```python
plt.rcParams['figure.figsize']=(5,5)
plt.bar(ss_df['Ship Mode'],ss_df['Sales']);
plt.rcParams.update({'font.size':3});
plt.xlabel('Ship Mode');
plt.ylabel('Sales');
```

# ***Here we observe that the sales are high if the ship mode is standard class, and sales are low if the ship mode is either second class or same day.***

```python
sns.countplot(x=ss_df.Segment)
```

# ***In the Home Office Segment, we observe that the count is less. So convergent strategy and improvement is a necessity for this Segment***

```python
plt.figure(figsize=(10,8))
sns.heatmap(ss_df.corr(),cmap='Blues',annot=True);
```

# ***Sales and Profit are Moderately Correlated,***
# ***Discount and Profit are Negatively Correlated***

```python
sns.lineplot(x='Sub-Category',y='Profit',label='In Trending',data=ss_df)
plt.legend()
plt.show()
```

# ***Here we can see that in Sub-Category, Machines gain more profit than others.***

```python
plt.rcParams['figure.figsize']=(28,16)
plt.bar(ss_df['Sub-Category'],ss_df['Discount']);
plt.rcParams.update({'font.size':14});
plt.xlabel('Sub_Category');
plt.ylabel('Discount');
```

# ***In this case, Binders,Appliances,Bookcases,Machines provided maximum discount***

```python
plt.rcParams['figure.figsize']=(20,5)
plt.rcParams.update({'font.size':12})
sns.countplot(x='Sub-Category',data=ss_df)
plt.show()
```

# ***From the above data, its very much clear that the Copiers and Machines Subcategory needs improvement.***

```python
plt.figure(figsize=(15,5))
plt.title('Sales City Wise')
ss_df['City'].value_counts().head(10).plot.pie()
plt.show()
```

# ***From the above figure you can see that most sales city-wise happend in Los Angeles.***
